```python
plt.hist(df['Tour Duration'], bins=20, edgecolor='black')
plt.title('Distribution of Tour Duration')
plt.xlabel('Tour Duration (hours)')
plt.ylabel('Frequency')
plt.show()
```

```python
px.box(df,x='Site Name',y='Satisfaction',color='Site Name',template='plotly_dark',color_discrete_sequence=['cyan','magenta'])
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.barplot(x='Accessibility', y='Satisfaction', data=df)
plt.title('Satisfaction by Accessibility')
plt.show()
```

```python
sns.lineplot(x='System Response Time', y='Satisfaction', data=df)
plt.title('System Response Time vs Satisfaction')
plt.show()
```

```python
px.scatter(df, x="Tour Duration", y="Satisfaction", color="Tour Duration", title="Tour Duration vs Satisfaction",template='plotly_dark',color_discrete_sequence=['cyan','magenta'])
```

```python
df_pie = df['Accessibility'].value_counts().reset_index()
df_pie.columns = ['Accessibility', 'count']
px.pie(df_pie, names='Accessibility', values='count', title="Accessibility Distribution",labels=True,template='plotly_dark',color_discrete_sequence=px.colors.qualitative.Plotly)
```

```python
sns.countplot(x='Site Name', hue='Tourist Rating', data=df)
plt.title('Tourist Rating Distribution by Site')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.pairplot(df[['Tour Duration', 'System Response Time', 'Recommendation Accuracy', 'Satisfaction']])
plt.show()
```

```python
df_grouped = df.groupby(['Tourist Rating', 'Accessibility']).size().unstack().fillna(0)
df_grouped.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Satisfaction by Tourist Rating and Accessibility')
plt.xlabel('Tourist Rating')
plt.ylabel('Count')
plt.show()
```
